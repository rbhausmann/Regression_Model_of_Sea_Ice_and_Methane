```python
def get_Nino(year):
    Nino_years=Nino_Nina_Data.El_Nino.values
    Nino_years_2 = []
    for year_ in Nino_years: 
        try: 
            Clean_year = year_.replace('early','').replace('mid','').replace('late','').replace(' ','')
            Nino_split= Clean_year.split('-')
            for year_2 in Nino_split:
                Nino_years_2.append(int(year_2))
        except: 
            pass
    if year in Nino_years_2:
        return 1
    else: 
        return 0
    
```

```python
AGAGE_1986['El_Nino'] = AGAGE_1986['Year'].apply(get_Nino)
```

```python
def get_Nina(year):
    Nina_years=Nino_Nina_Data.La_Nina.values
    Nina_years_2 = []
    for year_ in Nina_years: 
        try: 
            Clean_year = year_.replace('early','').replace('mid','').replace('late','').replace(' ','')
            Nina_split= Clean_year.split('-') 
            for year_2 in Nina_split:
                Nina_years_2.append(int(year_2))
        except: 
            pass
    if year in Nina_years_2:
        return 1
    else: 
        return 0
```

```python
AGAGE_1986['La_Nina'] = AGAGE_1986['Year'].apply(get_Nina)
```
